```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
from utils import kmeans

%matplotlib inline
```

```python
original_img = cv2.imread('images/colors.jpg')
original_img = cv2.cvtColor(original_img, cv2.COLOR_BGR2RGB)

print(original_img.shape)
img = np.reshape(original_img, (-1,3))
img = img/255
print(img.shape)
centroids, pixel_type = kmeans(img, K=3, iters=3)
```

```python
print(centroids)
segmented_img = centroids[pixel_type.flatten()]
segmented_img = segmented_img.reshape(np.array(original_img).shape)

plt.figure()
plt.subplot(1,2,1)
plt.imshow(original_img)
plt.subplot(1,2,2)
plt.imshow(segmented_img)
```
