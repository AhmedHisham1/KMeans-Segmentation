```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
from utils import kmeans

%matplotlib inline
```

```python
original_img = cv2.imread('images/the-pyramids.jpg')
original_img = cv2.cvtColor(original_img, cv2.COLOR_BGR2RGB)

print(original_img.shape)
img = np.reshape(original_img, (-1,3))
img = img/255
print(img.shape)
centroids_2, pixel_type_2 = kmeans(img, K=2, iters=10)
centroids_4, pixel_type_4 = kmeans(img, K=4, iters=12)
centroids_8, pixel_type_8 = kmeans(img, K=8, iters=15)
```

```python
segmented_img_2 = centroids_2[pixel_type_2.flatten()]
segmented_img_2 = segmented_img_2.reshape(np.array(original_img).shape)

segmented_img_4 = centroids_4[pixel_type_4.flatten()]
segmented_img_4 = segmented_img_4.reshape(np.array(original_img).shape)

segmented_img_8 = centroids_8[pixel_type_8.flatten()]
segmented_img_8 = segmented_img_8.reshape(np.array(original_img).shape)

fig = plt.figure(figsize=(40,20))
plt.subplot(2,2,1)
plt.imshow(original_img)
plt.title('Original Image', fontsize=30)
plt.subplot(2,2,2)
plt.imshow(segmented_img_2)
plt.title('K=2', fontsize=30)
plt.subplot(2,2,3)
plt.imshow(segmented_img_4)
plt.title('K=4', fontsize=30)
plt.subplot(2,2,4)
plt.imshow(segmented_img_8)
plt.title('K=8', fontsize=30)
```
